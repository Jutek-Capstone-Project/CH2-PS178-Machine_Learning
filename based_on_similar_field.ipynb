# PCA for seeing the correlation for each item

```python
pca_field = field_dataset.drop(['no','id', 'name', 'vicinity', 'kecamatan', 'kabupaten', 'user_ratings_total'], axis = 1)
pca_field
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(pca_field), columns = pca_field.columns)
X_scaled
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

n_components = 2  # Choose the number of components you want
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)
```

```python
from yellowbrick.cluster import KElbowVisualizer
# Quick examination of elbow method to find numbers of clusters to make.
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(X_pca)
Elbow_M.show()
```

```python
n_clusters = 4  # Number of clusters you want to find
kmeans = KMeans(n_clusters=n_clusters)
```

```python
kmeans.fit(X_pca)
cluster_labels = kmeans.labels_
```

```python
cluster_labels
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, cmap='viridis')
plt.title('PCA with Clustering')
plt.legend()
plt.show()
```

```python
from math import cos, asin, sqrt, pi, sin, radians, atan2
def recommend_similar_field(field, current_lat, current_lng):
  row_condition = field_dataset['id'] == field
  row_index = field_dataset.index[row_condition][0]

  # Check which columns have a value of 1 for the specified row
  columns_with_1 = field_dataset.iloc[row_index, 1:] == 1

  columns = ['jenis_sepakbola', 'jenis_badminton', 'jenis_tenis', 'jenis_futsal', 'jenis_voli', 'jenis_basket']
  for i in columns:
    if(columns_with_1[i] == True):
      jenis = i
      break

  pca_field = field_dataset[['lat', 'lng', 'rating', 'price', i, 'fasilitas_wifi', 'fasilitas_parkir_motor', 'fasilitas_parkir_mobil', \
                             'fasilitas_wc', 'fasilitas_kantin', 'fasilitas_mushola']]

  pca_field = pca_field[pca_field[i] == 1]

  scaler = StandardScaler()
  X_scaled = pd.DataFrame(scaler.fit_transform(pca_field), columns = pca_field.columns, index = pca_field.index)
  X_scaled

  n_components = 2  # Choose the number of components you want
  pca = PCA(n_components=n_components)
  X_pca = pca.fit_transform(X_scaled)

  n_clusters = 4  # Number of clusters you want to find
  kmeans = KMeans(n_clusters=n_clusters)
  kmeans.fit(X_pca)
  cluster_labels = kmeans.labels_
  pca_field['cluster'] = kmeans.predict(X_pca)

  # row_condition = pca_field['id'] == field
  # row_index = pca_field.index[row_condition][0]

  cluster = pca_field.iloc[row_index]['cluster']

  rec_index = pca_field.index[pca_field['cluster'] == cluster].tolist()
  list_recommend = field_dataset.iloc[rec_index]

  def distance(lat1, lon1, lat2, lon2):
    R = 6371.0 # kilometer

    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c

    return distance

  list_recommend['dist'] = None
  list_recommend['dist'] = list_recommend.apply(lambda x: distance(current_lat, current_lng, x['lat'], x['lng']), axis=1)

  return list_recommend.sort_values(by=['dist'], ascending=True).iloc[0:5]

```
